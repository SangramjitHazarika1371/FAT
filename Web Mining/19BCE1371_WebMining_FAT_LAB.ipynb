## Question

### Write a hybrid web crawler program which takes as input a url, search word and maximum number of pages to be searched and starts with breadth-first- search to be followed by depth-first-search and returns as output all the web pages it searched till it found the search word on a web page or return failure.

Website Base Link Taken: https://en.wikipedia.org/wiki/Web_mining

Text to be found: "web" or any input text passed in the function (Here web is taken as an example)

Maximum number of pages to be searched is taken as input

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import urllib.request
from bs4 import BeautifulSoup
import re
```

Breadth First Search (BFS)

```python
from urllib.request import urljoin
from bs4 import BeautifulSoup
import requests
from urllib.request import urlparse
  
links_intern = set()
input_url = "https://en.wikipedia.org/wiki/Web_mining"
depth = 1
  
links_extern = set()
  
def level_crawler(input_url):
    temp_urls = set()
    current_url_domain = urlparse(input_url).netloc

    beautiful_soup_object = BeautifulSoup(
        requests.get(input_url).content, "html.parser")

    for anchor in beautiful_soup_object.findAll("a"):
        href = anchor.attrs.get("href")
        if(href != "" or href != None):
            href = urljoin(input_url, href)
            href_parsed = urlparse(href)
            href = href_parsed.scheme
            href += "://"
            href += href_parsed.netloc
            href += href_parsed.path
            final_parsed_href = urlparse(href)
            is_valid = bool(final_parsed_href.scheme) and bool(
                final_parsed_href.netloc)
            if is_valid:
                if current_url_domain not in href and href not in links_extern:
                    print("External Website Link - {}".format(href))
                    links_extern.add(href)
                if current_url_domain in href and href not in links_intern:
                    print("Internal Website Link - {}".format(href))
                    links_intern.add(href)
                    temp_urls.add(href)
    return temp_urls

def bfs(): 
    if(depth == 0):
        print("Internal Website Link - {}".format(input_url))
    elif(depth == 1):
        level_crawler(input_url)
    else:
        queue = []
        queue.append(input_url)
        for j in range(depth):
            for count in range(len(queue)):
                url = queue.pop(0)
                urls = level_crawler(url)
                for i in urls:
                    queue.append(i)
```

Depth First Search (DFS)

```python
pages = set()
find = 0
def getLinks(pageUrl, max_links, text):
    global pages
    if len(pages)<max_links:
        global find
        with urllib.request.urlopen('https://en.wikipedia.org{}'.format(pageUrl)) as url:
            s = url.read()
            find = s.find(text.encode())
        html = urlopen('https://en.wikipedia.org{}'.format(pageUrl))
        wordlist = []
        bsObj = BeautifulSoup(html.read(), 'html.parser')
        if find == -1:
            print('https://en.wikipedia.org{}'.format(pageUrl))
            print ("Failure!! Word Not Found")
        else:
            print('https://en.wikipedia.org{}'.format(pageUrl))
            print("Hey!!! 'web' Word is found")
            print("Search completed since the word 'web' is found")
            
        if find == -1: 
            for link in bsObj.find("div", {"id":"bodyContent"}).findAll("a", href=re.compile("^(/wiki/)((?!:).)*$")):
                if 'href' in link.attrs:
                    if link.attrs['href'] not in pages and len(pages)<max_links:
                        newPage = link.attrs['href']
                        print('-'*20)
                        pages.add(newPage)
                        getLinks(newPage, max_links, text)

def dfs():
    no_of_links = int(input("Enter the number of urls you want the crawler to crawl in order to search the word 'web': "))
    getLinks('/wiki/Web_mining', no_of_links, 'web')
    
```
